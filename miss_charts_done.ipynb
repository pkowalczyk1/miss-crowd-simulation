```python
import pandas as pd
import csv
```

```python
with open("/content/output.csv", newline="") as infile, open(
    "/content/output_cleaned.csv", "w", newline=""
) as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    for row in reader:
        filled_row = []
        last_value = ""
        for cell in row:
            if cell == "":
                filled_row.append(last_value)
            else:
                last_value = cell
                filled_row.append(cell)
        writer.writerow(filled_row)
```

```python
df_raw = pd.read_csv("/content/output_cleaned.csv")
```

```python
df_raw.head(10)
```

```python
df_raw=df_raw.T
df_raw.head(10)
```

```python
df_raw = df_raw.reset_index(drop=False)
df_raw.columns = df_raw.iloc[0]
df_raw.head(10)
```

```python
df_raw = df_raw.drop(0)
```

```python
df_raw['[run number]'] = df_raw['[run number]'].apply(lambda x: x.split('.')[0])
```

```python
df_raw=df_raw.rename(columns={"[run number]": "run-number",
                              "[reporter]": "reporter",
                              "[final]": "final",
                              "[min]": "min",
                              "[max]": "max",
                              "[mean]": "mean",
                              "[total steps]": "total-steps"})
df_raw=df_raw.replace("[step]", "step")
df_raw.head(10)
```

```python
print(df_raw.dtypes)
```

```python
df_raw['run-number'] = df_raw['run-number'].astype(int)
df_raw['fire-start'] = df_raw['fire-start'].astype(str)
df_raw['map-version'] = df_raw['map-version'].astype(str)
df_raw['turtles-cnt'] = df_raw['turtles-cnt'].astype(int)
df_raw['fire-speed'] = df_raw['fire-speed'].astype(int)
df_raw['fire-graceful-period'] = df_raw['fire-graceful-period'].astype(int)
df_raw['final'] = df_raw['final'].astype(int)
df_raw['min'] = df_raw['min'].astype(int)
df_raw['max'] = df_raw['max'].astype(int)
df_raw['mean'] = df_raw['mean'].astype(float)
df_raw['reporter'] = df_raw['reporter'].astype(str)

print(df_raw.dtypes)
```

```python
for reporter in ('step', 'safe-cnt', 'dead-cnt'):
  for metric in ('final', 'min', 'max', 'mean', 'total-steps'):
    df_raw[f'{reporter}_{metric}'] = df_raw.apply(lambda row: row[metric] if row['reporter'] == reporter else None, axis=1)
```

```python
df_raw = df_raw.drop(columns=['reporter', 'final', 'min', 'max', 'mean', 'total-steps'])
```

```python
df_raw.head(50)
```

```python
def fill_none(group):
    return group.ffill().bfill()

df_grouped = df_raw.groupby('run-number').apply(fill_none, include_groups=False)
```

```python
df_grouped = df_grouped.reset_index(drop=True)
```

```python
df_grouped.head(10)
```

```python
df = df_grouped.drop_duplicates()
df = df.reset_index(drop=True)
print(df.size, df_grouped.size)
df.head()
```

```python
df.drop(columns=['safe-cnt_min',
                 'safe-cnt_max',
                 'safe-cnt_mean',
                 'safe-cnt_total-steps',
                 'dead-cnt_max',
                 'dead-cnt_mean',
                 'dead-cnt_total-steps',
                 'dead-cnt_min',
                 'step_min',
                 'step_max',
                 'step_mean',
                 "step_total-steps"], inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Define filter parameters
speed = 2
graceful_period = 5

filtered = df[(df['fire-speed'] == speed) & (df['fire-graceful-period'] == graceful_period)].copy()
filtered['safe_percent'] = (filtered['safe-cnt_final']+1)/ filtered['turtles-cnt'] * 100
for fire_start in filtered['fire-start'].unique():
    subset = filtered[filtered['fire-start'] == fire_start]
    grouped = subset.groupby(['turtles-cnt', 'map-version'])['safe_percent'].mean().reset_index()
    plt.figure(figsize=(10, 6))
    for map_version, group in grouped.groupby('map-version'):
        plt.plot(group['turtles-cnt'], group['safe_percent'], marker='o', label=f'Map: {map_version}')

    plt.title(f'Safe % | Fire Start: {fire_start} | Speed={speed}, Grace Period={graceful_period}')
    plt.xlabel('Total agents count')
    plt.ylabel('Safe %')
    plt.legend(title='Map Version')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
